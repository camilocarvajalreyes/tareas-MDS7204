```python
from scipy.signal import lombscargle, periodogram
import numpy as np
from plot import plot_series, plot_spectrum
```

```python
hr1 = np.loadtxt('hr.7257.txt')
hr2 = np.loadtxt('hr.11839.txt')

plot_series([hr1,hr2],['pulso 1','pulso 2'],title='Series de pulso sanguineo')
```

```python
f, _ = periodogram(hr1)
```

```python
Pxx_den1 = lombscargle(np.array(range(len(hr1))),hr1,f[1:])
Pxx_den2 = lombscargle(np.array(range(len(hr2))),hr2,f[1:])
```

```python
plot_spectrum(f[1:],[Pxx_den1,Pxx_den2],['pulso 1','pulso 2'],max_freq=0.1)
```
