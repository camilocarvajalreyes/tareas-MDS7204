# Tarea 2 - Análisis de señales

```python
from scipy.signal import periodogram
import numpy as np
from plot import plot_series, plot_spectrum
```

```python
N = int(2**10)

omega = 10*2*np.pi
cosenos = lambda t: np.sum([n*np.cos(n*omega*t) for n in range(1,6)])
t_max = 0.5

serie_cos = np.array([cosenos(t) for t in np.linspace(0,t_max,N)])
```

```python
plot_series(serie_cos,'suma de cosenos',fig_size=(8,5))
```

```python
f, Pxx_den = periodogram(serie_cos)
plot_spectrum(f,Pxx_den,'suma de cosenos',log=False,fig_size=(8,5),max_freq=0.1,colour='tab:purple')
```
